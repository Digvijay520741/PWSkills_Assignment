      Q.6 Explain the concept of generators in Python and how they are defined?

       -- >>  A generator is a special type of function in Python that
             returns an iterator and  yields value one at a time instead of returning them all at once
                    It's mainly used for:

              Saving memory (does not store all data in memory)

              Lazy evaluation (values are generated only when needed)
              Working with infinite or large sequences

             Generators are defined just like normal functions, but they use the yield keyword instead of return.
            Imagine you want to process 1 billion numbers â€” storing
            them in a list would consume a lot of memory.
                              
            Ex: -
            def fib(n):
                a = 0
                b = 1
          for i in range(n): #if n=5, 0, 1, 2, 3, 4
              yield a
              a, b = b, a+b
         f = fib(10000000000000)
        f
        output: <generator object fib at 0x7f7ffa642340>

      Q.10  What is the difference between â€œmap()', 'reduce()", and â€œfilter()"
            functions in Python?

            -- >>  map () - Transform each element
                                      
               Purpose: Apply a function to every element in an iterable and return a new  iterable with the results.
              Use when: You want to transform or modify each element

              Syntax: -

                map(function, iterable)
                                 


               filter() - Keep only certain elements

              Purpose: Filter elements based on a condition (True/False) and return only the elements that meet the condition.

              Use when: You want to select a subset of elements.

              Syntax: -

                filter(function, iterable)
                                
            reduce() - Combine all elements into one value

          Purpose: Repeatedly apply a function to the iterable to reduce it to a  single result.

          Use when: You want a single output from a sequence (like sum, product,
                    max, etc.)

          reduce() is not built-in â€” you import it from functools:

          Syntax:-
            reduce(function, iterable)

          Ex:-  Using All Three Together(map, filter, return )
                from functools import reduce

          numbers = [1, 2, 3, 4, 5, 6]

          #  map: double each number
          mapped = list(map(lambda x: x * 2, numbers))
          print(mapped)

          # filter: keep only numbers > 5
          filtered = list(filter(lambda x: x > 5, mapped))
          print(filtered)

         # reduce: sum them all
         result = reduce(lambda x, y: x + y, filtered)
         print(result)

```python
 # 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

def sum_even_num(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:   # check if the number is even
            total = total + num
    return total

num_lst = [4,7,8,4,2,5,7,6]
print("list is given :", num_lst)
result_even = sum_even_num(num_lst)
print("Total sum of even no. in given list :" , result_even)
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def list_sq(num):
    num_sq = []
    for i in num:
        num_sq.append(i**2)
    return  num_sq
num_lst  = [2,7,5,3,7]
print("Given list:", num_lst)

result_sq = list_sq(num_lst)
print("square of given list :" , result_sq)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200

def prime_num(numbers):
     if numbers < 2:
          return False
     for i in range (2, numbers):
          if numbers % i == 0:
                return False
     return True
print("Total prime number between 1 and 200 : " )
for n in range(1, 201):
    if prime_num(n):
        print( n, end = " ")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci_seq(num):
    a = 0
    b = 1
    for i in range(num):
        yield a
        a , b = b , a + b

fib_series = fibonacci_seq(10)   # Generate first 10 Fibonacci numbers
print("Fibonacci series:")
for num in fib_series:
    print(num, end=" ")
```

```python
# 6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_two(n):
    for i in range(n):
        yield i, i ** 2  # produces one value at a time
n = 5
# Using the generator

for  i, val in power_two(n):
    #print(val, end=" ")
    print(f"power of" ,{i}, "is : " ,{val})
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    """Generator function that yields one line at a time from a file."""
    with open(file_path, 'r') as file:  # open file in read mode
        for line in file:
            yield line.rstrip('\n')  # remove newline character at the end .  removes only spaces, tabs, and newlines (any whitespace) from the right end of a string.

# Example usage:
file_path = 'C:\\Users\\ER DIGVIJAY KUMAR\\Downloads\\python_file.txt'  # replace with your file path

for line in read_file_lines(file_path):
    print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Sample list of tuples
data = [("Ashish", 85), ("Bhushan", 72), ("Chandan", 90), ("Digvijay",17),("Sweta", 14)]

# Sort using lambda (second element is at index 1)
sorted_list = sorted(data, key=lambda x: x[1])

print("Original list:", data)
print("Sorted list (by second element):", sorted_list) # tuple is sorte  with second element with ascending order
```

```python
# 9. Write a Python program that uses "map()" to convert a list of temperatures from Celsius to Fahrenheit.

cel_temp = [0,37, 96 , 100 , 20, 14 ]

#
#  Fahreheit: (C Ã— 9/5) + 32
fah_temp = list(map(lambda c: (c * 9/5) + 32, cel_temp))

# Print the results
print("Celsius temperatures:", cel_temp)
print("Fahrenheit temperatures:", fah_temp)
```
