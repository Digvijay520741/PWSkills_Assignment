Q.1   Q.1 What is Simple Linear Regression (SLR)? Explain its purpose.

 -- >> Simple Linear Regression (SLR) is a statistical and machine-learning method used to understand and model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to the data.
      
      In simple word, Simple Linear Regression tries to find a straight line that best shows how Y changes when X changes.

The equation of Simple Linear Regression is:

                               Y=mX+c
Where:
           Y = Dependent variable (output)
          X = Independent variable (input)
          m = Slope (how much Y changes when X increases by 1)
          c = Intercept (value of Y when X = 0)

Purpose of Simple Linear Regression
The main purposes of SLR are:

1.	To understand the relationship:
It helps us see whether X and Y are positively related, negatively related, or not related.

2. To predict values:
Once the line is fitted, we can predict Y for a given value of X.

3. To measure the effect of X on Y:
The slope (m) tells us how much Y will increase or decrease when X changes.

4. To explain trends:
It helps explain trends in data in a simple and clear way.





 Q.2 Q.2 What are the key assumptions of Simple Linear Regression?
           
  -- >> The key assumptions of Simple Linear Regression (SLR) are the conditions that should be satisfied for the model to give reliable and accurate results.

1. Linearity

Assumption:

The relationship between X (independent variable) and Y (dependent variable) is linear.

Meaning:

When X increases or decreases, Y changes in a straight-line manner.

Example:

Study hours ↑ → Exam marks ↑ (in a straight line)

2. Independence of Errors

Assumption:

The errors (residuals) are independent of each other.

Meaning:

An error for one observation should not affect another.

Example:

Marks of one student should not depend on marks of another student.


3. Homoscedasticity

Assumption:

The variance of errors is constant for all values of X.

Meaning (simple):

The spread of errors should be roughly the same everywhere.

 Bad: Errors spread increases as X increases
 Good: Errors spread looks equal across X

4. Normality of Errors

Assumption:

The residuals (errors) are normally distributed.

Meaning:

Most errors are near zero, and large errors are rare.

Why important:

This helps in accurate confidence intervals and hypothesis testing.

5. No Perfect Multicollinearity

Assumption:

Since this is Simple Linear Regression, there should be only one independent variable.

Meaning:

X should not be a combination of other variables (this is more important in Multiple Linear Regression).

6. No Significant Outliers

Assumption:

There should be no extreme outliers that strongly influence the regression line.

Meaning:

One unusual data point should not change the entire line.


Q.3 Q.3 Write the mathematical equation for a simple linear regression model and explain each term.

-- >> Mathematical Equation of Simple Linear Regression
The simple linear regression (SLR) model is written as:

                 Y=β0+β1X+ε

Explanation of Each Term
  	Y → Dependent variable (Output)
      The value we want to predict or explain
      Example: Exam marks, salary, house price

 	X → Independent variable (Input)
  	The factor that influences Y
     Example: Study hours, years of experience
  
  	β0 → Intercept
     The value of Y when X = 0
  	It shows where the regression line cuts the Y-axis
     Example: Base salary when experience is zero
      	β1 → Slope (Regression coefficient)
     	Shows how much Y changes when X increases by 1 unit
      If β1 = 5: Y increases by 5 units for every 1-unit increase in X
•	ε → Error term (Residual)
o	The difference between actual and predicted values
o	Represents noise or factors not included in the model

Simplified Form (Without Error Term)
Sometimes written as:
Y^=β0+β1X
Y^ = Predicted value of Y

Example
If the equation is:
               Y=40+6X
•	Intercept (40): Marks when study hours = 0
•	Slope (6): Each extra hour of study increases marks by 6

Q.5 Q.5 What is the method of least squares in linear regression?

-- >> The method of least squares is a mathematical technique used to find the best-fitting straight line for a set of data points in linear regression.
          In simple word, The method of least squares chooses the regression line such that the sum of the squares of the errors (residuals) is minimum.

                Error=Actual value−Predicted value

Least Squares Objective Function
The goal is to minimize:
	                           n
                             ∑ (yi−y^i)2
                            i=1


Where:

•	yi= actual value

•	y^i= predicted value

•	n = number of observations

We square the errors to:

•	Avoid positive and negative errors canceling out

•	Give more penalty to large errors

Regression Line Using Least Squares

The simple linear regression line is:

                                  y^=β0+β1x

The least squares method finds the values of β0 and β1 that minimize the total
squared error.

Important of least squares in liner regression

•	Gives the best possible fit for data

•	Produces unique and optimal estimates

•	Easy to compute and widely used

•	Foundation of most regression techniques

Real-World Example

In predicting exam marks from study hours, least squares find the line that keeps the overall difference between actual marks and predicted marks as small as possible.

Q.7 Q.7 Name and briefly describe three common evaluation metrics for regression models.

  -- >> Three common evaluation metrics for regression models

    1. Mean Absolute Error (MAE)
Definition:
MAE is the average of the absolute differences between actual and predicted values.
                         MAE=1/n ∑∣yi−y^i∣
Meaning:
It tells us, on average, how far predictions are from actual values.
Key point:
•	Easy to understand
•	All errors are treated equally

2. Mean Squared Error (MSE)
Definition:
MSE is the average of squared differences between actual and predicted values.
MSE=1/n ∑(yi−y^i)2
Meaning:
Large errors are penalized more because of squaring.
Key point:
•	Sensitive to outliers
•	Commonly used in training models

3. Root Mean Squared Error (RMSE)
Definition:
RMSE is the square root of MSE.
RMSE=  sq. root (MSE)
Meaning:
Shows error in the same units as the target variable, making it easier to interpret.
Key point:
•	Penalizes large errors
•	More interpretable than MSE


Q.8  Q.8 What is the purpose of the R-squared metric in regression analysis?

-- >> Purpose of the R-squared (R²) Metric in Regression Analysis
R-squared (R²) is a statistical measure that tells us how well a regression model explains the variation in the dependent variable.

Mathematical Form
   R² =  1  -   (SS₍res₎/ SS₍tot₎)
Where:
•	SS₍res₎ = Sum of squared residuals (unexplained variation)
•	SS₍tot₎ = Total sum of squares (total variation in Y)
Value Range and Meaning
•	R² = 1 → Perfect fit (100% variation explained)
•	R² = 0 → Model explains nothing
•	R² = 0.75 → 75% of variation in Y is explained by the model

Purpose of R²
1.	Measures goodness of fit
o	Tells how well the regression line fits the data.
2.	Explains model effectiveness
o	Shows how much of the outcome is explained by predictors.
3.	Compares models (with same dataset)
o	Higher R² generally indicates a better-fitting model.
4.	Easy to interpret
o	Expressed as a percentage or proportion.

Simple Example
If R² = 0.80 in a model predicting exam marks from study hours:
•	80% of the variation in marks is explained by study hours
•	20% is due to other factors (sleep, health, stress)


```python
"""
Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
(Include your Python code and output in the code box below.)

"""

from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
# X = hours studied, y = exam score
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([40, 45, 50, 55, 60])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

Q.10 How do you interpret the coefficients in a simple linear regression model?

    -- >> Interpreting Coefficients in a Simple Linear Regression Model
A simple linear regression (SLR) model is written as:
            Y = β0+β1X+ ε
The coefficients β0 and β1 tell us how X affects Y.
Intercept (β0)
Meaning:
•	It is the expected value of Y when X = 0.
Interpretation:
When the independent variable is zero, the dependent variable will be β0\beta_0β0.
Example:
If
             Y=40+6X
•	Intercept = 40
•	When study hours = 0, expected marks = 40
Note:
Sometimes X = 0 has no real-world meaning, but the intercept is still important for the equation.

 Slope (β1)
Meaning:
•	It shows the change in Y for a one-unit increase in X.
Interpretation:
For every 1-unit increase in X, Y changes by β1\beta_1β1 units (on average).
Example:
If slope = 6:
•	Each extra hour of study increases marks by 6 points
Sign of the slope:
•	Positive (+) → Y increases as X increases
•	Negative (−) → Y decreases as X increases

Simple Real-Life Example
Model:
               Salary=25,000+5,000×(Years of Experience)
•	Intercept (25,000): Starting salary with zero experience
•	Slope (5,000): Salary increases by ₹5,000 for each additional year of experience
