Q .1  What are Type I and Type II errors in hypothesis testing, and how do
      they impact decision - making?

-- >>  Type I errors: - When we reject the null hypothesis in reality it was
       true, we make  a Type I error. It is denoted as Alpha (α).
                     In other word, we wrongly conclude that something
       has happened when actually nothing has happened.

       Example
       Suppose a COVID test says positive (infected), but the person is actually not infected.
       H₀: Person is NOT infected

       H₁: Person IS infected
          We reject H₀ (say person is infected), but actually H₀ was true → Type I error
      Impact on decision
      Leads to unnecessary actions

    	Example: Healthy person placed in quarantine

        Type II errors: - When we accept (fail to reject ) the null hypothesis, in reality it was false, we Make a Type II error. It is denoted as Beta (β).
                    In other word, we wrongly conclude nothing has happen
        when actually something has happened.

        Example

         COVID test says negative (not infected) but the person is actually infected.
         H₀: Person is NOT infected
         H₁: Person IS infected
         We accept H₀ (say person is not infected), but actually H₀ is false → Type II error
       
       Impact on decision
       Dangerously missing a real problem
       Example: Infected person continues normal activities → virus spread



 Q.2 What is the P-value in hypothesis testing, and how should it be
     interpreted in the context of the null hypothesis?

 -- >> In hypothesis testing, we check whether there is enough statistical
       evidence to reject a claim (called the null hypothesis).
       The P-value is one of the most important concepts used to make this
       decision.
         
       Def: - The P-value is the probability of getting a sample result as
              Extreme as, or more extreme than, the observed result, assuming  
       that the null hypothesis is true.
                       In other way we can say P-value measures the probability
       of observing our sample result, assuming that the null hypothesis is true.

      The P-value helps us decide: Whether to reject the null hypothesis (H₀) or fail to reject H₀ based on a chosen significance level (α), usually 0.05. P-value does NOT prove whether H₀ is true or false, it only measures evidence.
      If P-value ≤ α (e.g., 0.05)
      Evidence is strong against H₀
      Result is statistically significant
      We reject the null hypothesis
      
      If P-value > α
      Evidence is not strong enough
      Result is not statistically significant
      We fail to reject the null hypothesis

      Scenario:
      A company claims that the average life of a battery is 100 hours.
      We collect a sample and compute a P-value = 0.03 with α = 0.05.
      Interpretation
      P-value (0.03) < α (0.05)
      Probability of getting such sample data if the company's claim is true is only 3%.	Such a low probability shows the sample result is unlikely under H₀. So, we reject the null hypothesis.
      
     	Conclusion: The average battery life is not 100 hours (statistically significant difference).

 Q. 4  What is a confidence interval, and how does the margin of error
      influence its width and interpretation?


-- >>  A confidence interval (CI) is a range of values that is likely to
       contain an unknown population parameter (like the mean or proportion) with a certain level of confidence (such as 90%, 95%, or 99%).

       Think of it as:

       Point estimate ± Margin of Error
       
       For example:

       If the average height in a sample is 170 cm and the 95% CI is 170 ± 4, the interval is 166 to 174 cm. This means we are 95% confident that the true population average lies between 166 and 174 cm.

       Confidence Interval=Xˉ±Margin of Error
       
       The margin of error (ME) shows how much the sample estimate can vary from the true population value.

       Margin of Error= Z x σ/ root n
    
      It depends on:
       Confidence level (e.g., 95%, 99%)
       Standard deviation
       Sample size (n)

       Margin of Error Influences the Width of a Confidence Interval
       a. Larger Margin of Error → Wider CI

       Happens when:
       Confidence level is higher (e.g., 99%)
       Sample size is small
       Data has high variability
     
     	Interpretation: Less precise estimate

      b. Smaller Margin of Error → Narrower CI

      Happens when:
      Confidence level is lower (e.g., 90%)
      Sample size is large
      Variability is low
  	  Interpretation: More precise estimate
      Interpretation Example
     95% CI = 50 ± 3 → (47, 53)
     Margin of error = 3
     Narrower CI ⇒ More precise estimate
     95% CI = 50 ± 10 → (40, 60)
     Margin of error = 10
  	 Much wider ⇒ Less precise, more uncertainty
     


 Q. 5 Describe the purpose and assumptions of an ANOVA test. How does it extend
      hypothesis testing to more than two groups?

  -- >>  ANOVA is a statistical test used to determine whether the means of
         three or more groups are significantly different from each other. In other words, it is technique that helps compare three or more group means at once.

        we need ANOVA Test because:
       If you compare only two groups, you can use a t-test. But if you compare three or more groups, doing multiple t-tests increases the chance of a false result (Type-I error).  ANOVA solves this by comparing all group   means simultaneously in one test.

        Purpose of ANOVA:
       Check if there is a significant difference between the means of three or more groups.
       It tests:
       Null hypothesis (H₀): All group means are equal
       μ1=μ2=μ3=……
       
       Alternative hypothesis (H₁): At least one group mean is different.
       
       Simple Meaning: ANOVA tells us whether all groups are same or any one
       group is different.

       Assumptions of ANOVA
       ANOVA works correctly only when these assumptions are satisfied:
       Independence of observations
       Each data point must be independent (no repeated observations).
      
       Normality
       Data in each group should be approximately normally distributed.
     	 Homogeneity of variance (Equal variance
       All groups should have similar variance.
       σ1^2≈σ2^2≈σ3^2

      If variances are very different → ANOVA results may be invalid.

      Extension beyond two groups:
      A t-test compares two means.
      ANOVA compares k means (k ≥ 3) using one test.
     	It uses an F-statistic to evaluate whether the differences between group means are greater than expected by chance.
      If the ANOVA result is significant, it tells us:
      “At least one group is different.”
      But it does not tell which group is different.
      For that, we use post-hoc tests (Tukey, Bonferroni, etc.).
    
     Simple Analogy
     Imagine you want to compare the average marks of three classes:
     Class A
     Class B
     Class C
     Instead of doing A vs. B, B vs. C, A vs. C separately (which increases error),
   
     ANOVA checks all three at once.

```python
# Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

import numpy as np
from scipy.stats import norm

data = [52, 55, 49, 50, 53, 54, 48, 51] # sample data
population_mean = 50 #  μ
population_std = 10  # σ

sample_mean = np.mean(data) # cal sample mean
n = len(data)

z_value = (sample_mean - population_mean) / (population_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_value)))  # This gives the area in the upper tail.

print("Sample Mean:", sample_mean)
print("Z-value:", z_value)
print("P-value:", p_value)

alpha = 0.05 # Interpretation (significance level α = 0.05)

if p_value < alpha:
    print("Result: Reject the null hypothesis. Significant difference.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference.")

```

```python
# Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.



import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=1, size=100000)

sample_means = []            # store sample means
sample_size = 50             # n = 50 for each sample

for i in range(5000):        # take 5000 samples
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(population, bins=50)
plt.title("Original Population (Exponential = Non-normal)")

plt.subplot(1,2,2)
plt.hist(sample_means, bins=50)
plt.title("Distribution of Sample Means (CLT → Normal)")

plt.show()
```

```python
# Write a Python function to calculate and visualize the confidence interval for a sample mean.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_confidence_interval(data, confidence=0.95):

    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean

    margin = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    plt.figure(figsize=(8, 4))
    plt.errorbar(x=1, y=mean, yerr=margin, fmt='o', color='blue', capsize=10)
    plt.xlim(0, 2)
    plt.xticks([])
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()

sample_data = [12, 15, 14, 10, 13, 15, 16, 14, 13, 11]
plot_confidence_interval(sample_data)
```

```python
# Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.



from scipy.stats import chisquare

observed = [50, 30, 20]

expected = [40, 40, 20]

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", round(chi_stat, 4))
print("P-value:", round(p_value, 4))

alpha = 0.05 # Interpretation (significance value)

if p_value < alpha:
    print("❌ Significant difference found → Reject Null Hypothesis")
else:
    print("✔ No significant difference → Fail to Reject Null Hypothesis")
```
