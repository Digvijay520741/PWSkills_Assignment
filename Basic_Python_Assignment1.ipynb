   Q.7 What is the difference between "==" and "is" operators in python?
   -->> Equality operator (==)
        The equality operator (==) is used to compare the values of two objects.
        It returns True if the values are the same, and False if they are different.It does not check the memory location, only the contents.
        Ex:-
            x = 5
            y = 5
            print(x == y)   # True (both have value 5)

            a = [1, 2, 3]
            b = [1, 2, 3]
           print(a == b)   # True (values are same)

      Identity Operator (is)
        The identity operator (is) checks if two variables point to the exact same object in memory.
        It returns True if they are the same object, otherwise False.
        It doesn't care about values, only about object identity (memory location).
        a = [1, 2, 3]
        b = [1, 2, 3]
        print(a == b)   # True  → values are same
        print(a is b)   # False → not the same object in memory
        c = a
        print(a is c)   # True  → both point to same object

  Q.14 Describe a scenario where a while loop is more suitable than a for loop?
  -->> Scenario where a while loop is better:
      When you don't know in advance how many times you need to repeat something.
       EX:- Asking for correct password

             password = input("Enter password: ")
      print("Access Granted!")
        Here, we don't know how many times the user will type the wrong password.That's why while is better than for.

       Ex:- Waiting for user input to quit
       command = ""
      while command.lower() != "quit":
      command = input("Type a command (or 'quit' to exit): ")
       print("Program Ended")
       The loop runs until the user types quit.Again, we can't predict the exact number of repetitions.
       In summary , Use a while loop when the number of repetitions is unknown and depends on a condition (like user input, sensor reading, or system state).

```python
10 # Write code to demonstrate type casting with list elements.
data = ["10", "20.5", "Digvijay", "", "True", "False", 30, 40.7, "50"]
print("Original List:", data)

# ---- Cast to int (only if numeric) ----
int_list = []
for item in data:
    s = str(item)  # work with string version
    if s.replace(".", "", 1).isdigit():  # check if number (int/float style)
        int_list.append(int(float(s)))
    else:
        int_list.append(None)
print("Casted to int:", int_list)

# ---- Cast to float (only if numeric) ----
float_list = []
for item in data:
    s = str(item)
    if s.replace(".", "", 1).isdigit():
        float_list.append(float(s))
    else:
        float_list.append(None)
print("Casted to float:", float_list)

# ---- Cast to string (always works) ----
string_list = [str(item) for item in data]
print("Casted to string:", string_list)

# ---- Cast to bool (with 'True'/'False' handling) ----
bool_list = []
for item in data:
    s = str(item).lower()
    if s == "true":
        bool_list.append(True)
    elif s == "false":
        bool_list.append(False)
    else:
        bool_list.append(bool(item))  # default Python conversion
print("Casted to bool:", bool_list)
```

```python
14. # Write a program to reverse a string using a while loop.
# Take input from user
text = input("Enter a string: ")

# Initialize variables
reversed_text = ""
index = len(text) - 1   # start from last index

# Loop backwards using while
while index >= 0:
    reversed_text += text[index]   # add each character , reversed_text = reversed_text +  text[index]
    index -= 1                     # move one step left , index = index -1

print("Reversed string:", reversed_text)
```

```python
15. # Write a python program to calculate the factorial of a number provided by the user using a while loop.
number = int(input("Enter a number: "))
result = 1
i  = 1
while i <= number:
    result *= i   # result = result * i
    i += 1        # i = i +1

print("Factorial of", number, "is:", result)
```
