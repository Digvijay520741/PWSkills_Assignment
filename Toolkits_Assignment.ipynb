```python
# 3. Write a Python program that logs an error message to a log file when a division by zero exception occurs.


import logging

logging.basicConfig(
    filename='app.log',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)


def divide(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError as e:

        logging.error("Division by zero error occurred: %s", e)
        print(" Error: Division by zero! Check the log file for details.")


print("Starting division operation...")
divide(5, 0)

print("\nProgram finished. Check 'app.log' file for logged error details.")
```

```python
# 4. Write a Python program that reads from one file and writes its content to another file.

try:

    with open("input.txt", "r") as source_file:
        content = source_file.read()


    with open("output.txt", "w") as destination_file:
        destination_file.write(content)

    print("File copied successfully! Check 'output.txt'.")

except FileNotFoundError:
    print(" Error: Source file 'input.txt' not found.")
except Exception as e:
    print(f" An unexpected error occurred: {e}")
```

```python
# 5. Write a program that handles both IndexError and KeyError using a try-except block


numbers = [10, 20, 30]
student = {"name": "Rahul", "age": 21}

# First check for IndexError
try:
    print(numbers[5])
except IndexError:
    print(" IndexError: You are trying to access an index that does not exist in the list.")

# Then check for KeyError
try:
    print(student["grade"])
except KeyError:
    print("KeyError: You are trying to access a key that does not exist in the dictionary.")

finally:
    print(" Program execution completed.")
```

```python
# 8. Create a histogram using Seaborn to visualize a distribution.


import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
data = np.random.normal(loc=70, scale=10, size=100)  # mean=70, std=10

sns.histplot(data, bins=10, kde=True, color='skyblue')

plt.title("Distribution of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

plt.show()
```

```python
#. 10 Calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math': [85, 90, 78, 92, 88, 76, 95],
    'Science': [80, 85, 75, 89, 86, 70, 94],
    'English': [78, 82, 74, 88, 80, 72, 90],
    'Computer': [92, 95, 88, 98, 94, 85, 99]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

print(" Correlation Matrix:")
print(corr_matrix)

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Heatmap of Subjects")
plt.show()
```
