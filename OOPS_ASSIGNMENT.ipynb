  Q. 12 What is method overriding in OOP?

  -- >> Method Overriding is an OOP concept where a child class (subclass) provides its own implementation of a method that is already defined in the parent class (superclass).

 It allows a subclass to change or extend the behaviour of an inherited method without changing the parent class code

  Key Points:

  Method name and parameters must be exactly the same in both parent and child classes.
  It happens during runtime (so it’s also known as runtime polymorphism).
  Used to achieve different behavior for the same method in the child class.

  Real-Life Example:

  Think of a parent class Vehicle with a method start().
   A Car starts with a key.
   A Bike starts with a kick.

  Same method name start(), but different behavior in each subclass.
  
  Syntax:-

   class Vehicle:   # Parent class
        
        def start(self):
        print("Vehicle is starting...")

   class Car(Vehicle): # Child class 1
        
        def start(self):      # Overriding the parent class method
        print("Car starts with a key.")

   class Bike(Vehicle):  # Child class 2
        
        def start(self):      # Overriding the parent class method
        print("Bike starts with a kick.")
        
        # Creating objects
   v = Vehicle()

   c = Car()

   b = Bike()
     
     # Calling start() method from each object

   v.start()   # Output: Vehicle is starting...

   
   c.start()   # Output: Car starts with a key.
   
   b.start()   # Output: Bike starts with a kick.

   Q.24 What is method chaining in Python OOP?

      object. method1(). method2(). method3()

           def __init__(self, balance=0):

               print(f"Current Balance: ₹{self.balance}")
               return self

```python
# 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Child class 1
class Circle(Shape):
    def __init__(self, radii):
        self.radius = radii

    def area(self):
        return 3.14 * self.radius * self.radius

# Child class 2
class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.width = breadth

    def area(self):
        return self.length * self.width
c = Circle(7)
r = Rectangle(6,4 )

print("Area of Circle :", c.area())
print("Area of Rectangle :", r.area())
```

```python
# 10.Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y


    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)


    def __str__(self):           # By defining our own __str__() inside Vector, we are overriding the original one.
        return f"Vector({self.x}, {self.y})"

# Create two vector objects
v1 = Vector(520, 741)
v2 = Vector(808, 520)

# Add them using the overloaded '+' operator
result = v1 + v2

# Display result
print("Resultant Vector:", result)
```

```python
# 14.Create a class Employee with a method caIcuIate_saIary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.



class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)  # call parent constructor
        self.bonus = bonus

    # Overriding the calculate_salary method to include bonus
    def calculate_salary(self):
        base_salary = super().calculate_salary()  # get salary from parent class
        return base_salary + self.bonus

# Create Employee object
emp = Employee(520, 1176)
print("Employee Salary:", emp.calculate_salary())

# Create Manager object
mgr = Manager(808, 741, 7410)  # Rs7410 bonus
print("Manager Salary:", mgr.calculate_salary())
```

```python
# 16.Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):

    # Abstract method (must be implemented in child classes)
    @abstractmethod
    def sound(self):
        pass


# Derived class 1
class Cow(Animal):
    def sound(self):
        return "Moo Moo"


# Derived class 2
class Sheep(Animal):
    def sound(self):
        return "Baa Baa"


# Create objects
cow = Cow()
sheep = Sheep()

# Call their sound methods
print("Cow sound:", cow.sound())
print("Sheep sound:", sheep.sound())
```

```python
# 17.Create a class Book with attributes title, author, and year_pubIished. Add a method get_book_info() that returns a formatted string with the book's details.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    # Method to return formatted book information
    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Create an object of Book
book1 = Book("Rich Dad Poor Dad ", "Robert Kiyosaki", 1997)

# Display the book information
print(book1.get_book_info())
```

```python
# 18.Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.


# Base class
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    # Method to display house details
    def get_info(self):
        return f"Address: {self.address}, Price: ₹{self.price}"

# Derived class
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        # Call parent (House) constructor
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    # Method to display mansion details (extends parent method)
    def get_info(self):
        return f"Address: {self.address}, Price: Rs {self.price}, Rooms: {self.number_of_rooms}"

# Create an object of Mansion
m1 = Mansion("Salt Lake Area  sec-5 ", 15000000, 5)

# Display mansion details
print(m1.get_info())
```
