Q.2 Explain overfitting and underfitting in ML. How can you detect and prevent them?

-- >> Overfitting: - Overfitting happens when a model learns too much detail,
ncluding noise and random fluctuations in the training data.
As a result, it performs very well on training data but poorly on test data.

    Overfitting = Low Bias + High Variance
(Model learns too much → fails to generalize)
  
Overfitting detection:
 High training accuracy + Low testing accuracy
 Large difference between train and test performance
 Validation loss increases while training loss decreases

Overfitting Prevention:
 Use cross-validation
 Use regularization (L1/L2)
 Use dropout (for neural networks)
 Stop training early
 Simplify model

 Underfitting: - Underfitting happens when a model is too simple to learn the
underlying patterns in the data. It performs poorly on both training and testing data

    Underfitting = High Bias + Low Variance
  (Model is too simple → cannot learn patterns)

Underfitting Detection:
Low training accuracy
Low test accuracy
Training and testing errors are high and close

Underfitting Prevention:
Use a more complex model
Train longer
Reduce regularization
Add more features


Q.3 How would you handle missing values in a dataset? Explain at least three methods  with examples.

-- >> Three methods to handle missing values in a dataset are:
  1. Deletion Method (Removing missing data)
     When the number of missing values is very small, we can remove those rows or columns.

 Example: Dataset
                            Age                       Salary  
                             25                         40,000
                             NA                         35,000
                             30                           NA
Remove rows with missing values:
Result:
Age	Salary
25	40,000
Python code:
df.dropna(inplace=True)

2. Mean/Median Imputation (Fill missing values)

This method replaces missing values with the mean or median of the column.
Use mean when data is normal (no outliers).
Use median when there are outliers.

Example:
Age column: [25, NaN, 35, 40]
Median = 35
Fill missing value with 35 → [25, 35, 35, 40]

Python code:
df['Age'].fillna(df['Age'].median(), inplace=True)

3. Predictive Modelling (Advanced method)

Here missing values are filled using machine learning models that predict the missing value based on other features.

Example:
If “Age” is missing, we can use features like Salary, Gender, Education to predict Age using:
Linear Regression
KNN Imputer
Decision Tree

Python code using KNN Imputer:
from sklearn.impute import KNNImputer

df = imputer.fit_transform(df)



Q.4 What is an imbalanced dataset? Describe two techniques to handle it (theoretical +  practical).


-- >> An imbalanced dataset is a dataset where one class has many more
      samples than the other class. This causes the model to become biased toward the majority class.

Example:

Fraud detection dataset:
98% = non-fraud
2% = Fraud

The model may learn to always predict “non-fraud” and still get high accuracy, which is misleading.

Two Techniques to Handle Imbalanced Data

SMOTE (Synthetic Minority Oversampling Technique):

SMOTE creates new synthetic samples for the minority class by interpolating between existing minority samples.
Unlike simple duplication, SMOTE generates slightly different data points, improving model learning.

Use Case:
Useful when minority class is very small
Works well for classification tasks like fraud detection, churn prediction

Python code:
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_res, y_res = sm.fit_resample(X, y)

2. Random Under/Oversampling

Random Oversampling: Increase the minority class by duplicating existing
                    minority samples. Useful when the dataset is small.

Python code:

 from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X, y)


Random Under sampling: Reduce the majority class by removing some majority
                      samples. Useful when majority class is very large.



Python code:

from sklearn.linear_model import LogisticRegression

Q.5 Why is feature scaling important in ML? Compare Min-Max scaling and
  Standardization.

 -- >> Feature scaling means bringing all features to a similar range.
          It is very important because:

Distance-based algorithms like KNN, SVM, K-Means depend on distance.
If one feature has a big range, it will dominate the distance calculation.

Example:
Age: 18-60
Salary: 10,000-1,00,000
→ Salary has a very large scale → model focuses more on salary → wrong prediction

Gradient Descent becomes faster
When features are in different scales, the cost function takes a zig-zag path → slow training.
With scaling → smooth & fast convergence.

Improves model accuracy & stability
Makes learning easier for ML algorithms.


Types of Feature Scaling:

Min-Max Scaling (Normalization)
Formula:
             X' = (X - Xmin)/ (Xmax - Xmin)

Output range: 0 to 1
 Best for: Neural Networks, KNN, K-Means
 Sensitive to outliers
Example:
Marks: 20, 30, 50 (Min = 20, Max = 50)
30= (30-20)/(50-20) = 10/30 = 0.33

Standardization (Z-Score Scaling) Formula:
                  X'=(X-μ)/σ

→ Output: Mean = 0, Standard deviation = 1
→ Range: Not fixed, can be negative or >1

 Works well even with outliers

Best for: SVM, Logistic Regression, Linear Regression
(Algorithms that assume data is normally distributed)

Example:
Heights (cm): 150, 160, 170
Mean = 160, SD = 8.16 approx

170 (170-160)/8.16 ≈1.22

Comparison Table
Feature Scaling Method	Output Range	Sensitive to Outliers?	Best Used In
Min-Max Scaling	0 to 1	Yes	KNN, K-Means, Neural Networks
Standardization	Mean=0, SD=1	No (less sensitive)	SVM, Linear Models

Conclusion
Feature scaling keeps all features in similar range → better distance calculation, faster gradient descent, and improved model performance.




Q.6 Compare Label Encoding and One-Hot Encoding. When would you prefer one over
  the other?

 -- >> Label Encoding
     Definition: Converts categorical values into integers. Each unique category is assigned a unique numeric label.
 Preferable situation: Use for ordinal variables where the order matters (e.g., Small < Medium < Large).
Example:

Screenshot 2025-12-03 125337.png


Pros:
Simple and memory-efficient.
Works well with ordinal data (where categories have a meaningful order).

Cons:
Imposes an artificial order on nominal data (categories without order).
Some algorithms may interpret the numbers as having magnitude (e.g., linear regression, distance-based models like KNN), which can mislead the model.
Use Case:
Ordinal categorical variables, e.g., Low, Medium, High.

One-Hot Encoding
Definition: Converts categorical values into binary vectors. Each category becomes a separate column, with 1 indicating presence and 0 indicating absence.
Preferable situation: Use for nominal variables where there is no natural order (e.g., Red, Green, Blue).


Example:


Screenshot 2025-12-03 125529.png

Pros:
Avoids introducing any ordinal relationship.
Safe for nominal categorical variables.

Cons:
Increases dimensionality (especially with high-cardinality features).
Can lead to sparse matrices, which may slow down some algorithms.
Use Case:

Nominal categorical variables, e.g., Color, Country, Gender.

```python
# Google Play Store Dataset
# a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings,
# and what could be the possible reasons?

# Google Play Store: Category vs Rating analysis
# Run in Jupyter / Colab / local Python (requires pandas, numpy, matplotlib, seaborn)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
# Put the csv in the same folder or provide full path
df = pd.read_csv("googleplaystore.csv")

# 2. Quick look
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
display(df.head())

# 3. Check data types & missing values
print("\nData types:\n", df.dtypes)
print("\nMissing values per column:\n", df.isnull().sum())

# 4. Basic cleaning & conversions
# Convert 'Rating' to numeric (some bad rows might turn to NaN)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Clean 'Reviews' (sometimes string) and convert to numeric
if 'Reviews' in df.columns:
    df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Clean 'Installs' by removing characters like '+' and ',' then convert
if 'Installs' in df.columns:
    df['Installs_clean'] = df['Installs'].astype(str).str.replace('[+,]', '', regex=True)
    df['Installs_clean'] = pd.to_numeric(df['Installs_clean'], errors='coerce')

# Clean 'Price' (remove $ sign)
if 'Price' in df.columns:
    df['Price_clean'] = df['Price'].astype(str).str.replace('[$]', '', regex=True)
    df['Price_clean'] = pd.to_numeric(df['Price_clean'], errors='coerce')

# Clean 'Size' - many formats: '19M', '1,000k', 'Varies with device'
def size_to_kb(s):
    try:
        s = str(s).strip()
        if s == 'Varies with device' or s == '' or s.lower() == 'nan':
            return np.nan
        if s.endswith('M'):
            return float(s[:-1]) * 1024  # convert MB to KB
        if s.endswith('k'):
            return float(s[:-1])
        # some entries may be exact numbers (bytes) or have commas
        s2 = s.replace(',', '')
        return float(s2)
    except:
        return np.nan

if 'Size' in df.columns:
    df['Size_KB'] = df['Size'].apply(size_to_kb)

# 5. How many ratings are missing?
print("\nRatings missing before drop:", df['Rating'].isna().sum())

# For this analysis we will drop rows where Rating is NaN because we need rating to compute averages
df_r = df.dropna(subset=['Rating']).copy()
print("After dropping NaN ratings shape:", df_r.shape)

# 6. Detect outliers in 'Rating' (ratings are typically 0-5 but check)
print("\nRating stats:\n", df_r['Rating'].describe())

# If you want to detect outliers in 'Reviews' or 'Installs_clean' use IQR:
def iqr_filter(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return lower, upper

# Example: outlier thresholds for Reviews (if present)
if 'Reviews' in df_r.columns:
    rev_lower, rev_upper = iqr_filter(df_r['Reviews'].dropna())
    print(f"\nReviews IQR bounds: lower={rev_lower}, upper={rev_upper}")
    # number of extreme review outliers
    print("Extreme review outliers:", ((df_r['Reviews'] < rev_lower) | (df_r['Reviews'] > rev_upper)).sum())

# 7. Optionally remove rows with extreme review counts or installs - here we keep all but note them
# For category-rating averages we typically do not remove reviews outliers because rating itself is the target.

# 8. Group by Category - compute mean rating and count of apps per category
cat_stats = (
    df_r
    .groupby('Category')['Rating']
    .agg(['mean', 'count', 'median', 'std'])
    .reset_index()
    .rename(columns={'mean':'avg_rating','count':'num_apps'})
)

# 9. To avoid unreliable small-sample means, filter categories with at least N apps (e.g., N=10)
N = 10
cat_stats_filtered = cat_stats[cat_stats['num_apps'] >= N].copy()

# 10. Sort by average rating
cat_stats_sorted = cat_stats_filtered.sort_values(by='avg_rating', ascending=False)

# Print top and bottom categories
print("\nTop categories by average rating (>= {} apps):".format(N))
display(cat_stats_sorted.head(10))

print("\nBottom categories by average rating (>= {} apps):".format(N))
display(cat_stats_sorted.tail(10))

# 11. Visualize (bar plots)
plt.figure(figsize=(10,6))
sns.barplot(data=cat_stats_sorted, y='Category', x='avg_rating', order=cat_stats_sorted['Category'])
plt.title('Average Rating by Category (categories with >= {} apps)'.format(N))
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.xlim(0,5)
plt.tight_layout()
plt.show()

# 12. Also show counts to judge reliability
plt.figure(figsize=(10,6))
sns.barplot(data=cat_stats.sort_values('num_apps', ascending=False).head(20), x='num_apps', y='Category')
plt.title('Top 20 Categories by Number of Apps')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.tight_layout()
plt.show()

# 13. Save results
cat_stats_sorted.to_csv("category_rating_summary.csv", index=False)
print("\nSaved category_rating_summary.csv")

# 14. Example of extracting highest and lowest category names
highest = cat_stats_sorted.iloc[0]
lowest = cat_stats_sorted.iloc[-1]
print("\nHighest avg rating category (filtered):", highest['Category'], "Avg Rating:", round(highest['avg_rating'],3), "Num apps:", int(highest['num_apps']))
print("Lowest avg rating category (filtered):", lowest['Category'], "Avg Rating:", round(lowest['avg_rating'],3), "Num apps:", int(lowest['num_apps']))
```

```python
# hr_analysis_simple.py
# Simple HR analytics: correlation, plots, and answering "more projects => more likely to leave?"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns        # optional but makes plots nicer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


df = pd.read_csv('hr_analytics.csv')

# 2) Quick look at the data
print("Dataset shape:", df.shape)
print(df.columns.tolist())
print(df.head())

# 3) Basic cleaning & conversions
salary_map = {'low': 0, 'medium': 1, 'high': 2}
if 'salary' in df.columns:
    df['salary_level'] = df['salary'].map(salary_map)

# Make sure boolean-like columns are numeric
for col in ['Work_accident', 'left', 'promotion_last_5years']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# 4) Numeric correlation with attrition (left)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if 'left' in numeric_cols:
    corr_with_left = df[numeric_cols].corr()['left'].sort_values(ascending=False)
    print("\nCorrelation with 'left' (sorted):")
    print(corr_with_left)
else:
    raise KeyError("'left' column not found in numeric columns")

# 5) Visualizations showing key drivers

plt.figure(figsize=(6,4))
sns.boxplot(x='left', y='satisfaction_level', data=df)
plt.title('Satisfaction level by Attrition (0=Stayed, 1=Left)')
plt.xlabel('Left (1=Yes)')
plt.ylabel('Satisfaction level')
plt.tight_layout()
plt.show()

# Average number_project by left (barplot)
proj_by_left = df.groupby('left')['number_project'].mean().reset_index()
plt.figure(figsize=(5,4))
sns.barplot(x='left', y='number_project', data=proj_by_left)
plt.title('Average number of projects: stayed vs left')
plt.xlabel('Left (1=Yes)')
plt.ylabel('Average number of projects')
plt.tight_layout()
plt.show()

# Salary vs attrition (counts)
plt.figure(figsize=(6,4))
if 'salary' in df.columns:
    sns.countplot(x='salary', hue='left', data=df, order=['low','medium','high'])
    plt.title('Salary level by attrition (counts)')
    plt.xlabel('Salary')
    plt.ylabel('Count')
    plt.legend(title='Left')
    plt.tight_layout()
    plt.show()

# Average monthly hours distribution by left (boxplot)
plt.figure(figsize=(6,4))
sns.boxplot(x='left', y='average_montly_hours', data=df)
plt.title('Average monthly hours by Attrition')
plt.xlabel('Left (1=Yes)')
plt.ylabel('Average monthly hours')
plt.tight_layout()
plt.show()

# Scatter: satisfaction vs monthly hours colored by left (gives visual sense of interaction)
plt.figure(figsize=(6,5))
sns.scatterplot(x='satisfaction_level', y='average_montly_hours', hue='left', data=df, alpha=0.6)
plt.title('Satisfaction vs Monthly Hours (colored by Left)')
plt.xlabel('Satisfaction level')
plt.ylabel('Average monthly hours')
plt.tight_layout()
plt.show()

# 6) Are employees with more projects more likely to leave?
proj_stats = df.groupby('number_project').agg(
    count=('left', 'size'),
    left_count=('left', 'sum')
).reset_index()
proj_stats['left_rate'] = proj_stats['left_count'] / proj_stats['count']
print("\nProj summary (count, left_count, left_rate):")
print(proj_stats.sort_values('number_project'))

# 7) Simple logistic regression to check the effect of number_project controlling for others
# Choose features that matter and are numeric: satisfaction_level, average_montly_hours, time_spend_company, number_project, salary_level
features = ['satisfaction_level', 'average_montly_hours', 'time_spend_company', 'number_project']
if 'salary_level' in df.columns:
    features.append('salary_level')

# Drop rows with missing values for chosen features + target
model_df = df[features + ['left']].dropna()

X = model_df[features].values
y = model_df['left'].values

# Standardize numeric features for coefficients interpretability
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# train/test split (just to fit a stable model)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42, stratify=y)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Coefficients (log-odds)
coef_df = pd.DataFrame({
    'feature': features,
    'coef': model.coef_[0]
}).sort_values(by='coef', key=abs, ascending=False)

print("\nLogistic regression coefficients (higher positive -> higher odds of leaving):")
print(coef_df)

# Convert coefficients to approximate odds ratios
coef_df['odds_ratio'] = np.exp(coef_df['coef'])
print("\nOdds ratios (exp(coeff)):")
print(coef_df[['feature','odds_ratio']])

# Model accuracy (optional)
acc_train = model.score(X_train, y_train)
acc_test = model.score(X_test, y_test)
print(f"\nModel accuracy: train={acc_train:.3f}, test={acc_test:.3f}")

# 8) Interpretations printed for convenience
print("\nQuick interpretations:")
print("- If 'satisfaction_level' has a large negative coefficient, lower satisfaction increases leaving probability.")
print("- A positive coefficient for 'number_project' means more projects (holding other factors equal) is associated with higher odds to leave.")
print("- Odds ratios > 1 mean higher odds of leaving; < 1 mean lower odds.")

# End of script
```
