  	Q.1 What are data structures, and why are they important?

     -->>  Data structure is a way of organizing, storing, and managing data
          so that it can be accessed and manipulated efficiently.
          In other way, Data structures is just like container (e.g., box, shelf, folder) that decides how data is arranged and how operations are performed on it.

   Examples of data structures:

  Primitive types: int, float, char, boolean

   Non-primitive:
  Linear structures: Arrays, Lists, Stacks, Queues

Non-linear structures: Trees, Graphs

Hash-based structures: Hash tables, Dictionaries, Sets

Data structures matter because how you organize data affects performance.

Key reasons:

 Efficiency: The right structure speeds up operations (e.g., search in array = O(n), but in BST = O(log n)).

Memory Use: They help manage memory effectively (e.g., linked lists grow dynamically, unlike fixed arrays).

Reusability: Well-tested structures can be applied across different projects.

Problem-Solving: Complex problems (navigation, scheduling, social networks) need specialized structures like tree, graphs or heaps.

Algorithm Support: Most algorithms depend on efficient data structures to work effectively.

  Q.4  Describe how dictionaries store data?

     -- >>  Dictionaries store data in key-value pairs using a hash table, which makes access and updates very fast.
         A dictionary stores data as key-value pairs. Each key is unique and acts like an address to its value.
                       Internally, Python uses a hash table:

                 The key is passed through a hash function.

               The hash function gives a number (hash value).
               That number decides where the key-value pair will be placed in memory (in a slot/box).

               Hash function: - Hashing makes searching very fast (average time = O(1)).

               Instead of checking each key one by one, Python directly jumps to the location using the hash.

                Ex: -
                        student = {
                                        "name": "Rahul",
                                          "age": 20,
                                          "course": "CS}
                 Explanation: - Step 1: Hash Function Generates Numbers

                                hash("name")  suppose 101

                                hash("age")  suppose 202

                                hash("course")  suppose 303

                                print(hash("name")) # Example output: 3120195363435791

                Step 2:   Hash Values  Map to Slots in Hash Table

               Think of the dictionary as a table of slots (like boxes in
               memory). Keys are converted into indexes using the hash value.                                                                      
               Step 3: Lookup
               print(student["age"]) # Python does: Compute hash("age")  202
                                     # Go directly to slot 202
              Return the value 20

      Q.8  What is a hash table, and how does it relate to dictionaries in Python?

      -->> A hash table is a data structure (like a smart container) that stores datain key- value pairs. It uses a hash function to convert a key into a number(Hash value). This hash value decides where the keyvalue pair will be placed in memory (inside a slot, also called a bucket), Because of this, searching, inserting, and deleting data is very fast (on average O(1) time).

      Example Dictionary
        In Python, a dictionary (dict) is built using a hash table internally.

        Keys  like locker numbers (must be unique).

        Values  what you keep inside the locker.

        Fast to store and find data.

                student = {
                          "name": "Rahul",
                          "age": 20,
                          "course": "CS"
                                          }

          Step 1: Hash Function Generates Numbers.

          hash("name")  suppose 101

          hash("age")  suppose 202

          hash("course")  suppose 303

          Step 2: Hash Values  Map to Slots in Hash Table.

                        Keys are converted into indexes using the hash value.
             Hash Table (slots in memory)

             Index 100 |  empty
             Index 101 | ("name", "Rahul")
             Index 102 |  empty
             Index 202 | ("age", 20)
             Index 303 | ("course", "CS")
             Index 304 |  empty

         Step 3: Lookup.

          When you call:

        print(student["age"])

        Python does:

   	       Compute hash("age")  202

  	      Go directly to slot 202

         Return the value 20

         That's why lookups are so fast.

  Collision Example
          If "roll" also hashes to 202, Python uses open addressing (finds next empty slot):
       Index 202 | ("age", 20)
       Index 203 | ("roll", 101)
        So, both can live in the dictionary without overwriting each other.

      Q.13 How do sets handle duplicate values in Python?

         -- >>  A set is an unordered collection of unique elements. If you try to add a duplicate value, Python automatically ignores it  only one copy is kept.

        How sets work internally

        Sets use a hash table.

         When you add an element, Python computes its hash value.

        If the hash already exists, the duplicate is ignored.

        Result  sets always keep unique values only.

        Duplicate handling

        If you add the same element again, Python gets the same hash.
        Since it already exists, Python ignores it. So, the set keeps only one copy of each value.

         Ex: - numbers = {1, 2, 2, 3, 4, 4, 5}

          print(numbers)

          output: {1, 2, 3, 4, 5}

          Step-by-step:

         {1, 2, 2, 3, 4, 4, 5}  has duplicates (2 and 4).

         Python removes duplicates automatically.

           Q.16 What is a nested dictionary, and give on example of its use case?

           -- >> A nested dictionary in Python is simply a dictionary inside another dictionary.
           It is useful when you want to store hierarchical or structured data (like JSON, configuration files, or databases).

          # Nested dictionary example
          students = {
          "101": {"name": "Alice", "age": 20, "course": "Python"},
          "102": {"name": "Bob", "age": 22, "course": "Java"},
          "103": {"name": "Charlie", "age": 21, "course": "C++"}
                }

        # Accessing nested dictionary values
         print(students["101] ["name"])   # Output: Alice
         print(students["102] ["course"]) # Output: Java

         Use case:
         Storing structured data (like students, employees, or products)
         Configuration settings in applications
        JSON-like data (API responses are often nested dictionaries)

       Q.19  Why are dictionaries considered unordered, and how does that affect data retrieval?

         -- >> A dictionary in Python stores data as key-value pairs.
             Dictionaries are logically unordered because theyre optimized for fast key access, not sequence order. This doesnt affect retrieving by key, but it means you shouldnt rely on iteration order (except in modern Python where order is preserved but not the primary feature).

             How this affects data retrieval

            Key-based retrieval is always consistent:

            my_dict = {"a": 1, "b": 2, "c": 3}
            
            print(my_dict["b"])  # Always 2

           You don't care about order, just about the key.
